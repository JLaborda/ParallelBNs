```python
def bold_extreme_values(data, format_string="%.2f", max_=True):
    #print("Data start:")
    #print(data)
    #print("Data End.")
    #print(data.name)
    if max_[data.name]:
        extrema = data != data.max()
    else:
        extrema = data != data.min()
    bolded = data.apply(lambda x : "\\textbf{%s}" % format_string % x)
    formatted = data.apply(lambda x : format_string % x)
    return formatted.where(extrema, bolded) 
```

```python
col_show_max = { "SHD": False, "LL Score": True, "BDeu Score": True, "dfMM": False, "dfMM minus": False, "dfMM plus": False, "Total iterations": False, "Total time(s)": False}
```

```python
test = df_groupby.copy()
str_test = ""
for net in test.index.levels[0].unique():
    df_net = test.loc[(net)]
    df_net = df_net.apply(lambda data : bold_extreme_values(data, max_=col_show_max), axis=0)
    str_test += df_net.to_latex(longtable=True,escape=False, label="tab:avg-"+net+"-results", caption="Resultados promedios de la red " + net + ".")
print(str_test)
```

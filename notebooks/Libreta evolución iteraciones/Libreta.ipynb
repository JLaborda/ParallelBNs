```python
import pandas as pd
import numpy as np
```

```python
keys = ["Iteration:","fes search:","FES STAGE:","ALPHA ORDER:","COMPUTE ALPHA H2:",
        "TRANSFORM DAGS:","TRANSFORM():","CONSENSUS UNION:","FUSION.UNION():",
        "SCORE FUSION:","SCORE CURRENT:","FUSE.RUN:","REMOVE INCONSISTENCIES:",
        "FUSION FES STAGE:","-bes search:","-BES STAGE:"]

with open('run_experiments_Pablo.sh.o26619.37') as entrada:
#with open('run_experiments_Pablo.sh.o26619.63') as entrada:

    with open('salida.csv', 'w') as salida:
        anterior = ""
        while(True):
            linea = entrada.readline()
            if not linea:
                break
            
            if linea.find("-FUSION BES STAGE:") != -1:
                    salida.write(linea.split()[-1] + '\n')
            else:
                for key in keys:
                    if linea.find(key) != -1:
                        if (key == "FES STAGE:" and anterior == "SCORE CURRENT:"):
                            
                            salida.write("0,0,")
                        if ((key == "-bes search:" and anterior != "-BES STAGE:")
                                or (key == "fes search:" and anterior != "SCORE CURRENT:")
                                or (key != "-bes search:" and key != "fes search:")):
                            salida.write(linea.split()[-1] + ',')
                        anterior = key
                        break
        
```

```python
header = ["Iteración","FES Th 1","FES Th 2","FES Th 3","FES Th 4","FES Th 5","FES Th 6","FES Th 7",
          "FES Th 8","FES Stage","Alpha Order","Compute Alpha H2","Transform DAGS","Transform()",
          "Consensus Union","Fusion.Union()","Score Fusion","Score Current",
          "Fuse.Run()","Remove Inconsistencies","Fusion FES Stage","BES Th 1","BES Th 2","BES Th 3",
          "BES Th 4","BES Th 5","BES Th 6","BES Th 7","BES Th 8","BES Stage","Fusion BES Stage"]

pd.set_option('max_columns', None)

df = pd.read_csv("salida.csv",names=header)
df = df.set_index("Iteración")
df
```

# TIEMPOS POR ITERACIÓN

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import imageio

# frames between transitions
n_frames = 30

x = ["FES Th 1","FES Th 2","FES Th 3","FES Th 4","FES Th 5","FES Th 6","FES Th 7",
          "FES Th 8","FES Stage","Alpha Order","Compute Alpha H2","Transform DAGS","Transform()",
          "Consensus Union","Fusion.Union()","Score Fusion","Score Current",
          "Fuse.Run()","Remove Inconsistencies","Fusion FES Stage","BES Th 1","BES Th 2","BES Th 3",
          "BES Th 4","BES Th 5","BES Th 6","BES Th 7","BES Th 8","BES Stage","Fusion BES Stage"]
coordinates_lists = df.values

bg_color = '#FFFFFF'
bar_color = '#283F4E'
gif_name = 'bars'

print('Creating charts\n')
filenames = []
for index in np.arange(0, len(coordinates_lists)-1):
    y = coordinates_lists[index]
    y1 = coordinates_lists[index+1]    
    y_path = np.array(y1) - np.array(y)    
    for i in np.arange(0, n_frames + 1):
        y_temp = (y + (y_path / n_frames) * i)        # plot
        fig, ax = plt.subplots(figsize=(25, 11))
        ax.set_facecolor(bg_color)        
        plt.bar(x, y_temp, color=bar_color)
        #plt.ylim(0,80)        # remove spines
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)        # grid
        ax.set_title("Iteración "+str(index+1))
        ax.set_axisbelow(True)
        plt.xticks(rotation=45)
        ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.7)        # build file name and append to list of file names
        filename = f'images/frame_{index}_{i}.png'
        filenames.append(filename)
        
        # last frame of each viz stays longer
        if (i == n_frames):
            for i in range(5):
                filenames.append(filename)        # save img
        plt.savefig(filename, dpi=96, facecolor=bg_color)
        plt.close()
print('Charts saved\n')# Build GIF
print('Creating gif\n')
with imageio.get_writer(f'{gif_name}.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
print('Gif saved\n')
print('Removing Images\n')
# Remove files
for filename in set(filenames):
    os.remove(filename)
print('DONE')
```

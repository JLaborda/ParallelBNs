# Introducción

Esta libreta contiene las representaciones de cada medida realizada en los experimentos del repositorio  [parallelBNs](https://github.com/JLaborda/ParallelBNs). Las representaciones realizadas se centran en las puntuaciones obtenidas en función del tamaño de la base de datos original, la red original, el número de hilos usados y su correspondiente interleaving. Para aclarar lo que significan las columnas con el sufijo `_delta`, disponemos de la siguiente ecuación:
$$\Delta = Score_{experiment} - Score_{control}$$

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Plot Functions

```python
def plot_delta_score(df_delta, y, data_size, interleaving):
    fig = plt.figure(figsize=(16,9))
    df = df_delta.loc[(df_delta["Data Size (k)"] == data_size) & (df_alarm_delta["Interleaving"] == interleaving)]
    ax = sns.barplot(x ="Threads", y=y, hue="Algorithm", data=df)
    plt.title(y +  " SCORE " + str(data_size) + "k instances for " + str(interleaving) + " interleaving")
    return ax
```

```python
# Change file for where the delta_scores.csv is in your computer
file = "../experiments/local/clean/networks/alarm/delta_scores.csv"
df_alarm_delta = pd.read_csv(file)
df_alarm_delta.head()
```

```python
data_sizes = [10, 50]
interleavings = [5, 10, 15]
measurements = ["SHD_delta", "BDeu Score_delta", "dfMM_delta", "dfMM plus_delta", "dfMM minus_delta", "Total time(s)_delta"]

for data_size in data_sizes:
        for y in measurements:
            for interleaving in interleavings:
                plot_delta_score(df_alarm_delta, y, data_size, interleaving)
```

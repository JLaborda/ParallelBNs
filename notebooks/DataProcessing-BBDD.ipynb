```python
import pandas as pd
import numpy as np
```

```python
folder_path = "networks/BBDD/"
networks = ["alarm", "andes", "barley", "cancer", "child", "earthquake", "hailfinder",
            "hepar2", "insurance", "link", "mildew", "munin", "pigs", "water", "win95pts"]
middle = ".xbif"
endings1 = ["50000_.csv","50001_.csv", "50002_.csv", "50003_.csv", "50004_.csv", "50005_.csv",
           "50006_.csv", "50007_.csv", "50008_.csv", "50009_.csv"]
endings2 = ["_.csv", "50001246_.csv"]
```

```python
for net in networks:
    for ending in endings1:
        path = folder_path + net + middle + ending
        print(path)
    for ending in endings2:
        path = folder_path + net + middle + ending
        print(path)
```

## 50k and 10k data creation

```python
for net in networks:
    dfs = []
    dfs2 = []
    df50 = None
    df10 = None
    print("Processing " + net + "...")
    for ending in endings1:
        path = folder_path + net + middle + ending
        dfs.append(pd.read_csv(path))
    # Merging dataframes into one
    df50 = pd.concat(dfs)
    save_path = folder_path + net + "50k.csv"
    df50.to_csv(save_path, index=False)
    
    for ending in endings2:
        path = folder_path + net + middle + ending
        dfs2.append(pd.read_csv(path))
    df10 = pd.concat(dfs2)
    save_path = folder_path + net + "10k.csv"
    df10.to_csv(save_path, index=False)
    print("DONE")
```
